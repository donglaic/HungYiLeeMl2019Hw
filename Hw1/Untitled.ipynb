```python
import pandas as pd
```

```python
my_res = pd.read_csv('result.csv').value
```

```python
ref_res = pd.read_csv('result (copy).csv').value
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(16,10))

ax.plot(range(240),my_res,range(240),ref_res)
ax.legend(['my','ref'])
```

```python
fig, ax = plt.subplots(figsize=(16,10))

ax.plot(range(240),my_res - ref_res)
# ax.legend(['my','ref'])
```
